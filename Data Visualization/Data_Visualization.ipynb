### 1. Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('netflix_titles.csv')

# Basic styling
sns.set(style='whitegrid', palette='muted', font_scale=1.1)
plt.rcParams['figure.figsize'] = (10, 6)
```

### 2. Movie vs TV Show Distribution

```python
sns.countplot(x='type', data=df)
plt.title('Distribution of Content Type on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

### 3. Top 10 Countries with Most Titles

```python
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

### 4. Content Growth Over Time

```python
content_per_year = df['release_year'].value_counts().sort_index()
sns.lineplot(x=content_per_year.index, y=content_per_year.values)
plt.title('Growth of Netflix Content Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

### 5. Ratings Distribution

```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

### 6. Most Common Genres

```python
from collections import Counter

# Split genres and count
genres = df['listed_in'].dropna().apply(lambda x: x.split(', '))
genre_counts = Counter([g for sublist in genres for g in sublist])
top_genres = pd.DataFrame(genre_counts.most_common(10), columns=['Genre', 'Count'])

sns.barplot(x='Count', y='Genre', data=top_genres)
plt.title('Top 10 Most Common Netflix Genres')
plt.show()
```

### 7. Top Directors

```python
top_directors = df['director'].value_counts().head(10)
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

### 8. Word Cloud of Descriptions

```python
from wordcloud import WordCloud

text = " ".join(df['description'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Words in Netflix Descriptions')
plt.show()
```
